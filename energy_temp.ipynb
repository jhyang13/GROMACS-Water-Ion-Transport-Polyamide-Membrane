### Umbrella Sampling Energy, Temp Graph

```python
# Import dependencies
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import os
```

### Potential Energy

```python
# Set the file path to the Excel file containing the stress-strain data.
data_file = r'Z:/UW_Madison_Jiahui/projects/PA_Membrane/mem_moy/subhamoy_swiss_itp/licl/S1_EneMin/min-energy.xvg'

# Load data from the specified text file into a NumPy array.
# The 'comments' parameter tells the function to ignore lines that start with these characters.
data_array = np.loadtxt(data_file, comments=('#', '@'))

# Print the shape of the array.
print("Data shape:", data_array.shape)
```

```python
# Extract the first column as the x-axis data.
x = data_array[:, 0]

# Extract the second column as the y-axis data.
y = data_array[:, 1]

# Create a figure for the plot with a specific size.
plt.figure(figsize=(10, 6))

# Plot the x and y data.
plt.plot(x, y, label='Potential Energy', color='red')

# Add labels for the x and y axes.
plt.xlabel("Time (ps)")
plt.ylabel("Potential Energy (kJ/mol)")

# Add a title to the plot.
# plt.title("GROMACS Energies")

# Display the legend.
plt.legend()

# Display the plot.
plt.show()
```
