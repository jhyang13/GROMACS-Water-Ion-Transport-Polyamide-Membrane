### NVT density_z Graph

```python
# Import dependencies
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import os
```

```python
# Set the file path to the Excel file containing the stress-strain data.
data_file_mem = r'Z:/UW_Madison_Jiahui/projects/PA_Membrane/mem_moy/subhamoy_swiss_itp/model/groups_mem.xvg'
data_file_wat = r'Z:/UW_Madison_Jiahui/projects/PA_Membrane/mem_moy/subhamoy_swiss_itp/model/groups_wat.xvg'
data_file_sod = r'Z:/UW_Madison_Jiahui/projects/PA_Membrane/mem_moy/subhamoy_swiss_itp/model/groups_SOD.xvg'
data_file_cal = r'Z:/UW_Madison_Jiahui/projects/PA_Membrane/mem_moy/subhamoy_swiss_itp/model/groups_CLA.xvg'

# Load data from the specified text file into a NumPy array.
# The 'comments' parameter tells the function to ignore lines that start with these characters.
data_array_mem = np.loadtxt(data_file_mem, comments=('#', '@'))
data_array_wat = np.loadtxt(data_file_wat, comments=('#', '@'))
data_array_sod = np.loadtxt(data_file_sod, comments=('#', '@'))
data_array_cal = np.loadtxt(data_file_cal, comments=('#', '@'))

# Print the shape of the array.
print("Data shape mem:", data_array_mem.shape)
print("Data shape wat:", data_array_wat.shape)
print("Data shape sod:", data_array_wat.shape)
print("Data shape cal:", data_array_wat.shape)
```

```python
# Create a figure for the plot with a specific size.
plt.figure(figsize=(10, 6))

# Plot each data set with a different label and color
plt.plot(data_array_mem[:, 0], data_array_mem[:, 1], label='PA Membrane', color='blue')
plt.plot(data_array_wat[:, 0], data_array_wat[:, 1], label='Water', color='green')
plt.plot(data_array_sod[:, 0], data_array_sod[:, 1], label='Sodium', color='red')
plt.plot(data_array_cal[:, 0], data_array_cal[:, 1], label='Chloride', color='black')

# Add labels for the x and y axes.
plt.xlabel("z (nm)")
plt.ylabel("Density")
min_x, max_x = 0, 30
plt.xticks(np.arange(min_x, max_x+1, 2))

# Add a title to the plot.
# plt.title("GROMACS Energies Comparison")

# Display the legend.
plt.legend()

# Display the plot.
plt.show()
```
