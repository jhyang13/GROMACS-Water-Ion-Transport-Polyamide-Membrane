### Umbrella Sampling PMF - Histogram Graph

```python
# Import dependencies
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import os
```

```python
# Set the file path to the Excel file containing the stress-strain data.
data_file = r'Z:/UW_Madison_Jiahui/projects/PA_Membrane/mem_moy/subhamoy_swiss_itp/nacl_5Molar/cl/S6_re/umbrellaS_re/histo.xvg'

# Load data from the specified text file into a NumPy array.
data_array = np.loadtxt(data_file, comments=('#', '@'))

# Print the shape of the array.
print("Data shape:", data_array.shape)
```

```python
# Filtering condition: Select rows where the first element in the second dimension
condition = (data_array[:, 0] >= -10) & (data_array[:, 0] <= 10)

# Use the condition to filter rows.
filtered_data = data_array[condition]

# Print the shape of the filtered array.
print("Filtered Data shape:", filtered_data.shape)
```

```python
# Extract the first element of the first dimension as the x-axis data.
x = filtered_data[:, 0]

# Extract all elements of the second dimension as the y-axis data.
y = filtered_data[:, 1:]

# Create a figure for the plot with a specific size.
plt.figure(figsize=(10, 6))

# Iterate through each set of y-values and plot curves.
for i in range(y.shape[1]):
    plt.plot(x, y[:, i], label=f'Curve {i+1}')  # Plot curves and add labels for each curve.

# Add labels for the x and y axes.
plt.xlabel("Reaction Coordinate")
plt.ylabel("Count")

# Add a title to the plot.
plt.title("Na+ PAM(NaCl) k=3000")

# Display the plot.
plt.show()
```
