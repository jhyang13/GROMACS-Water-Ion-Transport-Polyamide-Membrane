### Umbrella Sampling PMF Graph - Single Ion

```python
# Import dependencies
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import os
```

```python
# Set the file path to the Excel file containing the stress-strain data.
data_file_na = r'Z:/UW_Madison_Jiahui/projects/PA_Membrane/mem_moy/subhamoy_swiss_itp/nacl_single/na-direction/S6_re_iact_periodic_calculated/umbrellaS_re/bsResult.xvg'

# Load data from the specified text file into a NumPy array.
# Load data from the specified text file into a NumPy array.
data_array_na = np.loadtxt(data_file_na, comments=('#', '@'))

# Print the shape of the array.
print("Data shape:", data_array_na.shape)
```

```python

# Filtering condition: Select rows where the first element in the second dimension
condition_na = (data_array_na[:, 0] >= -6) & (data_array_na[:, 0] <= 6)

# Use the condition to filter rows.
filtered_data_na = data_array_na[condition_na]

# Print the shape of the filtered array.
print("Filtered Data shape:", filtered_data_na.shape)
```

```python
# Extract the first column as the x-axis data.
x_na = filtered_data_na[:, 0]

# Extract the second column as the y-axis data.
y_na = filtered_data_na[:, 1]

# Extract the third column as the y-axis error data.
error_na = filtered_data_na[:, 2]

# Create a figure for the plot with a specific size.
plt.figure(figsize=(10, 6))

# Plot the x and y data for Na+.
plt.plot(x_na, y_na, label='Periodic Na+ Calculated IACT', color='red')

# Fill the area around the Na+ curve with error bar as a shadow.
plt.fill_between(x_na, y_na - error_na, y_na + error_na, color='red', alpha=0.2)

# Add labels for the x and y axes.
plt.xlabel("z (nm)")
plt.ylabel("PMF (kCal/mol)")

# Add a title to the plot.
# plt.title("GROMACS Energies")

# Display the legend.
plt.legend()

# Display the plot.
plt.show()
```

```python
# Set the file path to the Excel file containing the stress-strain data.
data_file = r'Z:/UW_Madison_Jiahui/projects/PA_Membrane/mem_moy/subhamoy_swiss_itp/nacl_single/na-direction/S6_re_all/umbrellaS_re/histo.xvg'

# Load data from the specified text file into a NumPy array.
data_array = np.loadtxt(data_file, comments=('#', '@'))

# Print the shape of the array.
print("Data shape:", data_array.shape)
```

```python
# Filtering condition: Select rows where the first element in the second dimension
condition = (data_array[:, 0] >= -6) & (data_array[:, 0] <= 6)

# Use the condition to filter rows.
filtered_data = data_array[condition]

# Print the shape of the filtered array.
print("Filtered Data shape:", filtered_data.shape)
```

```python
# Extract the first element of the first dimension as the x-axis data.
x = filtered_data[:, 0]

# Extract all elements of the second dimension as the y-axis data.
y = filtered_data[:, 1:]

# Create a figure for the plot with a specific size.
plt.figure(figsize=(10, 6))

# Iterate through each set of y-values and plot curves.
for i in range(y.shape[1]):
    plt.plot(x, y[:, i], label=f'Curve {i+1}')  # Plot curves and add labels for each curve.

# Add labels for the x and y axes.
plt.xlabel("Reaction Coordinate")
plt.ylabel("Count")

# Add a title to the plot.
plt.title("Periodic Na+ calculated IACT")

# Display the plot.
plt.show()
```
