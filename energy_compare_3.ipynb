```python
# Import dependencies
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import os
```

```python
# Set the file path to the Excel file containing the stress-strain data.
data_file_1 = r'Z:/UW_Madison_Jiahui/projects/PA_Membrane/mem_moy/subhamoy_swiss_itp/nacl_single/na-direction/5_7_113/rdf.xvg'
data_file_2 = r'Z:/UW_Madison_Jiahui/projects/PA_Membrane/mem_moy/subhamoy_swiss_itp/nacl_single/na-direction/3_0_155/rdf.xvg'
data_file_3 = r'Z:/UW_Madison_Jiahui/projects/PA_Membrane/mem_moy/subhamoy_swiss_itp/nacl_single/na-direction/0_2_439/rdf.xvg'

data_file_1_1 = r'Z:/UW_Madison_Jiahui/projects/PA_Membrane/mem_moy/subhamoy_swiss_itp/nacl_single/na-direction/5_7_113/rdf_cn.xvg'
data_file_2_2 = r'Z:/UW_Madison_Jiahui/projects/PA_Membrane/mem_moy/subhamoy_swiss_itp/nacl_single/na-direction/3_0_155/rdf_cn.xvg'
data_file_3_3 = r'Z:/UW_Madison_Jiahui/projects/PA_Membrane/mem_moy/subhamoy_swiss_itp/nacl_single/na-direction/0_2_439/rdf_cn.xvg'

# Load data from the specified text file into a NumPy array.
# The 'comments' parameter tells the function to ignore lines that start with these characters.
data_array_1 = np.loadtxt(data_file_1, comments=('#', '@'))
data_array_2 = np.loadtxt(data_file_2, comments=('#', '@'))
data_array_3 = np.loadtxt(data_file_3, comments=('#', '@'))

data_array_1_1 = np.loadtxt(data_file_1_1, comments=('#', '@'))
data_array_2_2 = np.loadtxt(data_file_2_2, comments=('#', '@'))
data_array_3_3 = np.loadtxt(data_file_3_3, comments=('#', '@'))

# Print the shape of the array.
print("Data shape:", data_array_1.shape)
print("Data shape:", data_array_2.shape)
print("Data shape:", data_array_3.shape)

print("Data shape:", data_array_1_1.shape)
print("Data shape:", data_array_2_2.shape)
print("Data shape:", data_array_3_3.shape)
```

```python
# Extract the first column as the x-axis data.
x_1 = data_array_1[:, 0]
x_2 = data_array_2[:, 0]
x_3 = data_array_3[:, 0]

# Extract the second column as the y-axis data.
y_1 = data_array_1[:, 1]
y_2 = data_array_2[:, 1]
y_3 = data_array_3[:, 1]

# Create a figure for the plot with a specific size.
plt.figure(figsize=(10, 6))

# Plot the x and y data.
plt.plot(x_1, y_1, label='z = -5.5', color='red')
plt.plot(x_2, y_2, label='z = -3.0', color='green')
plt.plot(x_3, y_3, label='z = -0.0', color='blue')

# Add labels for the x and y axes.
plt.xlabel("r (nm)")
plt.ylabel("g(r)")

# Add a title to the plot.
plt.title("RDF of Na+ Surrounded by Water")

# Display the legend.
plt.legend()

# Display the plot.
plt.show()
```

```python
# Extract the first column as the x-axis data.
x_1 = data_array_1[:, 0]
x_2 = data_array_2[:, 0]
x_3 = data_array_3[:, 0]

x_1_1 = data_array_1_1[:, 0]
x_2_2 = data_array_2_2[:, 0]
x_3_3 = data_array_3_3[:, 0]

# Extract the second column as the y-axis data.
y_1 = data_array_1[:, 1]
y_2 = data_array_2[:, 1]
y_3 = data_array_3[:, 1]

y_1_1 = data_array_1_1[:, 1]
y_2_2 = data_array_2_2[:, 1]
y_3_3 = data_array_3_3[:, 1]

# 创建图形和左侧轴
plt.figure(figsize=(10, 6))
ax1 = plt.gca()
ax1.set_xlim(0, 1)
ax1.set_ylim(0, 8)
ax1.set_xlabel("r (nm)")
ax1.set_ylabel("g(r)")

ax1.plot(x_1, y_1, label='z = -5.5', color='red')
ax1.plot(x_2, y_2, label='z = -3.0', color='green')
ax1.plot(x_3, y_3, label='z = -0.0', color='blue')

ax1.legend(loc='upper left')
ax1.set_title("Na+ Surrounded by Water")

# 创建共享x轴的右侧轴
ax2 = ax1.twinx()
ax2.set_ylabel("CN")
ax2.plot(x_1, y_1_1, label='CN z = -5.5', color='red', linestyle=':')
ax2.plot(x_2, y_2_2, label='CN z = -3.0', color='green', linestyle=':')
ax2.plot(x_3, y_3_3, label='CN z = -0.0', color='blue', linestyle=':')
ax2.legend(loc='upper right')

plt.show()
```
