### Umbrella Sampling PMF Graph

```python
# Import dependencies
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import os
```

```python
# Set the file path to the Excel file containing the stress-strain data.
data_file = r'Z:/UW_Madison_Jiahui/projects/PA_Membrane/mem_moy/subhamoy_swiss_itp/na/S6_re/umbrellaS_re/profile.xvg'

# Load data from the specified text file into a NumPy array.
data_array = np.loadtxt(data_file, comments=('#', '@'))

# Print the shape of the array.
print("Data shape:", data_array.shape)
```

```python
# Filtering condition: Select rows where the first element in the second dimension
condition = (data_array[:, 0] >= -6.5) & (data_array[:, 0] <= 6.4)

# Use the condition to filter rows.
filtered_data = data_array[condition]

# Print the shape of the filtered array.
print("Filtered Data shape:", filtered_data.shape)
```

```python
# Extract the first column as the x-axis data.
x = filtered_data[:, 0]

# Extract the second column as the y-axis data.
y = filtered_data[:, 1]

# Create a figure for the plot with a specific size.
plt.figure(figsize=(10, 6))

# Plot the x and y data.
plt.plot(x, y, label='Na+', color='red')

# Add labels for the x and y axes.
plt.xlabel("z (nm)")
plt.ylabel("PMF (kCal/mol)")

# Add a title to the plot.
# plt.title("GROMACS Energies")

# Display the legend.
plt.legend()

# Display the plot.
plt.show()
```

```python
# Set the file path to the Excel file containing the stress-strain data.
data_file_na = r'Z:/UW_Madison_Jiahui/projects/PA_Membrane/mem_moy/subhamoy_swiss_itp/nacl_concentration/na/S6_re/umbrellaS_re/profile.xvg'
data_file_cl = r'Z:/UW_Madison_Jiahui/projects/PA_Membrane/mem_moy/subhamoy_swiss_itp/nacl_concentration/cl/S6_re/umbrellaS_re/profile.xvg'
data_file_h2o = r'Z:/UW_Madison_Jiahui/projects/PA_Membrane/mem_moy/subhamoy_swiss_itp/nacl_concentration/h2o/S6_re/umbrellaS_re/profile.xvg'

# Load data from the specified text file into a NumPy array.
data_array_na = np.loadtxt(data_file_na, comments=('#', '@'))
data_array_cl = np.loadtxt(data_file_cl, comments=('#', '@'))
data_array_h2o = np.loadtxt(data_file_h2o, comments=('#', '@'))

# Print the shape of the array.
print("Data shape:", data_array_na.shape)
print("Data shape:", data_array_cl.shape)
print("Data shape:", data_array_h2o.shape)
```

```python
low = -7
up = 6.5

# Filtering condition: Select rows where the first element in the second dimension
condition_na = (data_array_na[:, 0] >= low) & (data_array_na[:, 0] <= up)
condition_cl = (data_array_cl[:, 0] >= low) & (data_array_cl[:, 0] <= up)
condition_h2o = (data_array_h2o[:, 0] >= low) & (data_array_h2o[:, 0] <= up)
# Use the condition to filter rows.
filtered_data_na = data_array_na[condition_na]
filtered_data_cl = data_array_cl[condition_cl]
filtered_data_h2o = data_array_h2o[condition_h2o]

# Print the shape of the filtered array.
print("Filtered Data shape:", filtered_data_na.shape)
print("Filtered Data shape:", filtered_data_cl.shape)
print("Filtered Data shape:", filtered_data_h2o.shape)
```

```python
# Extract the first column as the x-axis data.
x_na = filtered_data_na[:, 0]
x_cl = filtered_data_cl[:, 0]
x_h2o = filtered_data_h2o[:, 0]

# Extract the second column as the y-axis data.
y_na = filtered_data_na[:, 1]
y_cl = filtered_data_cl[:, 1]
y_h2o = filtered_data_h2o[:, 1]

# Assume the first value in the y-axis data for each ion is the baseline to subtract
baseline_na = y_na[3]
baseline_cl = y_cl[3]
baseline_h2o = y_h2o[3]

# Subtract the baseline value from the y-axis data to shift the starting point to 0
adjusted_y_na = y_na - baseline_na
adjusted_y_cl = y_cl - baseline_cl
adjusted_y_h2o = y_h2o - baseline_h2o

# Create a figure for the plot with a specific size.
plt.figure(figsize=(10, 6))

# Plot the x and adjusted y data for Na+.
plt.plot(x_na, adjusted_y_na, label='Na+', color='red')

# Plot the x and adjusted y data for Cl-.
plt.plot(x_cl, adjusted_y_cl, label='Cl-', color='green')

# Plot the x and adjusted y data for H2O.
plt.plot(x_h2o, adjusted_y_h2o, label='H2O', color='blue')

# Add labels for the x and y axes.
plt.xlabel("z (nm)")
plt.ylabel("Adjusted PMF (kCal/mol)")

# Display the legend.
plt.legend()

# Display the plot.
plt.show()
```
