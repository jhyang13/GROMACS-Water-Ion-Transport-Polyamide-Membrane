### Energy Compare Graph

```python
# Import dependencies
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import os
```

```python
# Set the file path to the Excel file containing the stress-strain data.
data_file_1 = r'Z:/UW_Madison_Jiahui/projects/PA_Membrane/mem_moy/subhamoy_swiss_itp/nacl_single/na-direction/average_coul_energies.txt'
data_file_2 = r'Z:/UW_Madison_Jiahui/projects/PA_Membrane/mem_moy/subhamoy_swiss_itp/nacl_single/na-direction/caught-output.txt'
data_file_3 = r'Z:/UW_Madison_Jiahui/projects/PA_Membrane/mem_moy/subhamoy_swiss_itp/nacl_single/na-direction-extral/average_coul_energies.txt'
data_file_4 = r'Z:/UW_Madison_Jiahui/projects/PA_Membrane/mem_moy/subhamoy_swiss_itp/nacl_single/na-direction-extral/caught-output.txt'

# Load data from the specified text file into a NumPy array.
# The 'comments' parameter tells the function to ignore lines that start with these characters.
data_array_1 = np.loadtxt(data_file_1, comments=('#', '@'))
data_array_2 = np.loadtxt(data_file_2, skiprows=3, usecols=(0, 1), comments=('#', '@'))
data_array_3 = np.loadtxt(data_file_3, comments=('#', '@'))
data_array_4 = np.loadtxt(data_file_4, skiprows=3, usecols=(0, 1), comments=('#', '@'))
data_array_4[:, 1] = -data_array_4[:, 1]

# Print the shape of the array.
print("Data shape:", data_array_1.shape)
print("First 3 rows of data_array_1:")
print(data_array_1[:3])  # Print the first three rows

print("Data shape:", data_array_2.shape)
print("First 3 rows of data_array_2:")
print(data_array_2[:3])  # Print the first three rows

print("Data shape:", data_array_3.shape)
print("First 3 rows of data_array_3:")
print(data_array_3[:3])  # Print the first three rows

print("Data shape:", data_array_4.shape)
print("First 3 rows of data_array_4:")
print(data_array_4[:3])  # Print the first three rows
```

```python
# Assuming you have already loaded the data into data_array_1 and data_array_2 as described
# Create dictionaries to map first column values to second column values for quick lookup
dict1 = {row[0]: row[1] for row in data_array_1}
dict2 = {row[0]: row[1] for row in data_array_2}

# Create a list to hold the merged data
merged_data = []

# Iterate over the keys (first column values) and values (second column values) in the first dictionary
for key, value in dict1.items():
    if key in dict2:
        # If the key also exists in the second dictionary, append a new list with both second column values
        merged_data.append([dict2[key], value])

# Convert the list of lists to a NumPy array
merged_data_array_1 = np.array(merged_data)

# Print the shape of the array
print("Shape of merged_data_array:", merged_data_array_1.shape)
print("First three rows of merged_data_array:")
print(merged_data_array_1[:3])
```

```python
# Assuming you have already loaded the data into data_array_1 and data_array_2 as described
# Create dictionaries to map first column values to second column values for quick lookup
dict3 = {row[0]: row[1] for row in data_array_3}
dict4 = {row[0]: row[1] for row in data_array_4}

# Create a list to hold the merged data
merged_data = []

# Iterate over the keys (first column values) and values (second column values) in the first dictionary
for key, value in dict3.items():
    if key in dict4:
        # If the key also exists in the second dictionary, append a new list with both second column values
        merged_data.append([dict4[key], value])

# Convert the list of lists to a NumPy array
merged_data_array_2 = np.array(merged_data)

# Print the shape of the array
print("Shape of merged_data_array:", merged_data_array_2.shape)
print("First three rows of merged_data_array:")
print(merged_data_array_2[:3])
```

```python
# Assuming merged_data_array_1 and merged_data_array_2 are defined and properly formatted
# Plotting
plt.figure(figsize=(10, 6))  # Set the figure size for better visibility

# Plot for merged_data_array_1
plt.plot(merged_data_array_1[:, 0], merged_data_array_1[:, 1], color='orange', marker='o', linestyle='-')
# Plot for merged_data_array_2
plt.plot(merged_data_array_2[:, 0], merged_data_array_2[:, 1], color='orange', marker='o', linestyle='-')

plt.xlim(-6, 6)

# Adding titles and labels
#plt.title('Line Plot of Merged Data Arrays')
plt.xlabel('z (nm)')
plt.ylabel('Coulomb Short Range Energy (kJ/mol)')

# Adding a legend to distinguish the plots
#plt.legend()

# Show the plot
plt.show()
```
